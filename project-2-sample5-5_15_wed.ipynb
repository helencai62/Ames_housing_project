```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
train = pd.DataFrame(pd.read_csv('train.csv'))
train.columns = [i.lower().replace(' ', '') for i in train.columns]
```

```python
#from Tableau measures
train.loc[:, ['1stflrsf', '2ndflrsf', 'bsmtunfsf', 'bsmtfinsf1', 'garagearea', 'garageyrblt',
             'grlivearea', 'lotfrontage', 'masvnrarea', 'openporchsf', 'overallqual', 'screenporch', 'totalbsmtsf', 'totrmsabvgrd', 'wooddecksf', 'yearbuilt']].dtypes
```

```python
X_train_num = train.loc[:, [#'id',
                            '1stflrsf', '2ndflrsf', 'bsmtunfsf', 'bsmtfinsf1', 'garagearea',
                            'garageyrblt', 'grlivearea', 'lotfrontage', 'masvnrarea', 
                            'openporchsf', 'overallqual', 'screenporch', 'totalbsmtsf', 
                            'totrmsabvgrd', 'wooddecksf', 'yearbuilt'
                            #'saleprice'
]]

y_train_num = train['saleprice']
```

```python
X_train_num.fillna(value = 0, inplace = True)
X_train_num.isnull().sum()
```

```python
X_train_num_train, X_train_num_test, y_train_num_train, y_train_num_test = train_test_split(X_train_num, y_train_num, test_size = 0.25)
print(X_train_num_train.shape, y_train_num_train.shape)
print(X_train_num_test.shape, y_train_num_test.shape)
```

```python
#numerical model
lr = LinearRegression()
lr.fit(X_train_num_train, y_train_num_train)
lr.score(X_train_num_train, y_train_num_train)
```

```python
lr.score(X_train_num_test, y_train_num_test)
```
