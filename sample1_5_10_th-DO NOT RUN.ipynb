```python
subset = train_df.loc[:,['Overall Qual', 
                'Garage Area', 
                'Garage Cars', 
                'Total Bsmt SF', 
                '1st Fl SF', 
                'Year Built',
                'Year Remod/Add', 
                'Full Bath', 
                'Garage Yr Blt', 
                'Mas Vnr Area', 
                'TotRms AbvGrd',
                'SalePrice']]

subset.head()
```

```python
#looking at subset of data with above 0.5 correlation with SalePrice
    #don't brute force and run everything --> energy expenditure

fig = plt.figure(figsize = (20, 10))
    
for i, col in enumerate(subset.columns):
    fig.add_subplot(3, 4, i+1)
    plt.scatter(subset[col], subset['SalePrice']);
    plt.xlabel(col)
    plt.ylabel('SalePrice')
```

```python
#one variable: '1st Fl SF' does not show up
    #let's check the data type --> float64 type but all nulls --> remove from feature set
subset.dtypes
    #check for nulls --> we did this at the beginning with .info() and should have done it while creating the subset
subset.isnull().sum()
subset = subset.drop('1st Fl SF', axis = 1)
subset.head()
```

```python
subset.dropna(inplace = True)
subset.isnull().sum()
```

```python
#we can also rename the columns to be more analysis friendly
    #lowercase, remove spaces
subset.columns = [i.lower().replace(' ','') for i in subset.columns]
subset.head()
```

```python
#Instantiate our first model / lin reg

from sklearn.linear_model import LinearRegression
features = [col for col in subset.columns if (col != 'saleprice' and col != 'filler')]
X = subset[features]
y = subset['saleprice']

lr = LinearRegression()
lr.fit(X, y)
lr.score(X, y)

display(features)
```

```python
import statsmodels.api as sm

#QUESTION: wouldn't we expect the R-squared output to be 0.82 (same as above)?
#adding filler for Beta-naught intercept and multiplying by 1
subset['filler'] = 1

model=sm.OLS(y, X)
result=model.fit()
result.summary()
```
