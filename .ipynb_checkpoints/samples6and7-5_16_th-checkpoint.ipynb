```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
train = pd.DataFrame(pd.read_csv('train.csv'))
train.columns = [i.lower().replace(' ', '') for i in train.columns]
```

```python
#looking at object type columns in dataset
len(list(train.select_dtypes(include = 'object').columns))
train.select_dtypes(include = 'object').dtypes

#from Tableau dimensions
X_train_obj = train.loc[:, [#'id',
                        'bsmtcond', 'centralair', 'condition1',
                        'condition2', 'electrical', 'exterqual',
                        'fireplacequ', 'foundation', 'garagefinish',
                        'garagequal', 'garagetype', 'kitchenqual',
                        'landcontour', 'landslope', 'lotconfig',
                        'lotshape', 'mszoning', 'roofmatl', 'roofstyle',
                        'saletype', 'street', 'utilities'
                        #, 'saleprice'
                       ]]

X_train_obj.head()
```

```python
print('before', X_train_obj.isnull().sum())
X_train_obj.fillna(value = 0, inplace = True)
print('\n\n\n', 'after', X_train_obj.isnull().sum())
```

```python
#set target variable
y_train_obj = train['saleprice']
```

```python
#use pd.get_dummies to convert into numbers
    #drop_first = False here b/c I want to look at the correlation of all columns next
X_train_obj = pd.get_dummies(X_train_obj, drop_first = False)
```

```python
#train-test-split training dataset - for good practice
X_train_obj_train, X_train_obj_test, y_train_obj_train, y_train_obj_test = train_test_split(X_train_obj, y_train_obj, test_size = 0.25)

#matches train-test-split set shapes that we saw in creating the numerical set
print(X_train_obj_train.shape, y_train_obj_train.shape)
print(X_train_obj_test.shape, y_train_obj_test.shape)
```

```python
#instantiate, fit and score a categorical linear model
lr = LinearRegression()
lr.fit(X_train_obj_train, y_train_obj_train)
lr.score(X_train_obj_train, y_train_obj_train)
```

```python
#model is performing well on train set
    #lower score on test than train so overfit model --> higher variance
    #model is really bad at predicting home saleprices
lr.score(X_train_obj_test, y_train_obj_test)
```

```python
#look at our entire dataset and see what the top correlated values of each categorical variable are
X_train_obj['saleprice'] = y_train_obj
X_train_obj_corr = X_train_obj.corr()[['saleprice']].sort_values(by = 'saleprice', ascending = False)

#plot rank by correlation between categorical variable value and saleprice
figure = plt.figure(figsize = (25, 50))
sns.heatmap(X_train_obj_corr.rank(ascending = True), annot = X_train_obj_corr, annot_kws={"size": 20});
sns.set(font_scale=1.1)
```

```python
#arbitrarily determined threshold: take variable values with higher / lower than +/- 0.5 correlation
    #on pos corr end:
        #kitchenqual_Ex
        #foundation_PConc
    #on neg corr end:
        #kitchenqual_TA
        #exterqual_TA

#made some exceptions and included a few on the fence out of curiosity
    #on pos corr end:
        #exterqual_Ex
        #exterqual_Gd
    #on neg corr end:
        #fireplacequ_0 --> was going to include but decided against this as fireplacequ had 1000 nulls and those were replaced by 0's during cleaning

#categorical_variable_value_set = ['kitchenqual_Ex', 'foundation_PConc', 'kitchenqual_TA', 'exterqual_TA', 'exterqual_Ex', 'exterqual_Gd']
```

```python
#reset categorical variable set to limited values
X_train_obj = X_train_obj.loc[:, [#'id',
                               'kitchenqual_Ex', 'foundation_PConc', 'kitchenqual_TA', 'exterqual_TA', 
                               'exterqual_Ex', 'exterqual_Gd'
                                #, 'saleprice'
]]
```

```python
#reset train-test-split on new X_train_obj
X_train_obj_train, X_train_obj_test, y_train_obj_train, y_train_obj_test = train_test_split(X_train_obj, y_train_obj, test_size = 0.25)

print(X_train_obj_train.columns)
#refit model
lr.fit(X_train_obj_train, y_train_obj_train)
lr.score(X_train_obj_train, y_train_obj_train)
```

```python
#model is not particularly accurate, HOWEVER model score on test and train are closer together
    #confidence that model is better at predicting than before when using the original X_train_obj of values for each categorical variable input
    #lower model score for test than train so model is overfit --> higher variance
lr.score(X_train_obj_test, y_train_obj_test)
```

```python
test = pd.DataFrame(pd.read_csv('test.csv'))
test.head()
```

```python
test.columns = [i.lower().replace(' ', '') for i in test.columns]

#returning same categorical variable values in test as in X_train_obj
X_test_obj = pd.get_dummies(test.loc[:, [#'id',
            'kitchenqual', 'foundation', 'exterqual'
            ]], drop_first = False)

X_test_obj = X_test_obj.loc[:, ['kitchenqual_Ex', 'foundation_PConc', 'kitchenqual_TA', 'exterqual_TA', 
                               'exterqual_Ex', 'exterqual_Gd']]

X_test_obj.columns = [i.lower() for i in X_test_obj.columns]
X_test_obj.head()
```

```python
predictions = lr.predict(X_test_obj)
```

```python
X_test_obj['Id'] = test['id']
X_test_obj['Saleprice'] = predictions
```

```python
sample6 = X_test_obj.loc[:, ['Id', 'Saleprice']]

sample6.set_index('Id', inplace = True)
sample6.sort_index(axis = 0, inplace = True)
sample6.head()
```

```python
sample6.to_csv('sample6_hc')
```
